```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("forecasts/Forecasts_DE_DNN_LEAR_ensembles.csv")
# parse index as datetime
df.index = pd.to_datetime(df.index)

df = df[["Real price", "DNN Ensemble", "LEAR Ensemble"]]

# last days
days = 5
df = df.tail(24 * days)

# iterate over columns and plot
for col in df.columns:
    plt.plot(df[col], label=col)

# x axis month
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Price (/MWh)")

plt.legend()
plt.show()

```
