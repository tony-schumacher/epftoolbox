```python
import pandas as pd
import numpy as np

# Load your csv file
df = pd.read_csv(
    "forecasts/Forecasts_DE_DNN_LEAR_ensembles.csv", parse_dates=True, index_col=0
)


def sMAPE(y_true, y_pred):
    return (
        100
        / len(y_true)
        * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
    )


df.head()
```

```python

smape_dnn = sMAPE(df['Real price'], df['DNN 4'])
smape_lear = sMAPE(df['Real price'], df['LEAR 1092'])

print(f'sMAPE for DNN: {smape_dnn}%')
print(f'sMAPE for LEAR: {smape_lear}%')
```
